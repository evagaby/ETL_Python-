# **FUNÇÃO PARA LIMPAR DADOS DA TABELA DE 2022**

```python

def limpardados22 (dados):
  dados = dados[(dados['country'] == 'Brazil')]
  dados = pd.DataFrame(dados)
  dados = dados.rename(columns={'rank_in_country':'Ranking_no_pais','rank_in_subregion':'Ranking_na_Subregiao',
                        'institution':'Instituicao', 'country_code':'Codigo_do_pais','country':'Pais','size':'Tamando_da_instituicao','focus':'Foco'
                        ,'research':'Pesquisa'})
  dados ['rank_2021'] = dados ['rank_2021'].str.replace('+','')
  dados ['rank_2022'] = dados ['rank_2022'].str.replace('+','')
  dados ['ar_rank'] = dados ['ar_rank'].str.replace('+','')
  dados ['er_rank'] = dados ['er_rank'].str.replace('+','')
  dados ['fsr_rank'] = dados ['fsr_rank'].str.replace('+','')
  dados ['cpf_rank'] = dados ['cpf_rank'].str.replace('+','')
  dados ['ifr_rank'] = dados ['ifr_rank'].str.replace('+','')
  dados ['isr_rank'] = dados ['isr_rank'].str.replace('+','')
  return dados

df22_semi=limpardados22(dados=df22)
df22_limpo=pd.DataFrame(df22_semi)
print(df22_limpo)
```

TRATAMENTO FINAL E MOVENDO DADOS PARA PASTA

```python
df22_semi=limpardados22(dados=df22)
df22_limpo=pd.DataFrame(df22_semi)
df22_limpo.to_csv('Dados_2022_QS.csv',index=False)
caminho_arquivo_csv = 'Dados_2022_QS.csv'
caminho_pasta_destino = '/content/drive/MyDrive/projetopucminas/ARQUIVOS_CSV'
shutil.move(caminho_arquivo_csv, caminho_pasta_destino)
```

# **FUNÇÃO PARA LIMPAR DADOS DA TABELA DE 2023**

```python
def limpardados23 (dados=df23):
  dados = dados[(dados['country'] == 'Brazil')]
  dados = pd.DataFrame(dados)
  dados = dados.rename(columns={'rank_in_country':'Ranking_no_pais','rank_in_subregion':'Ranking_na_Subregiao',
                        'institution':'Instituicao', 'country_code':'Codigo_do_pais','country':'Pais','size':'Tamando_da_instituicao','focus':'Foco'
                        ,'research':'Pesquisa'})
  dados ['rank_2022'] = dados ['rank_2022'].str.replace('+','')
  dados ['rank_2023'] = dados ['rank_2023'].str.replace('+','')
  dados ['ar_rank'] = dados ['ar_rank'].str.replace('+','')
  dados ['er_rank'] = dados ['er_rank'].str.replace('+','')
  dados ['fsr_rank'] = dados ['fsr_rank'].str.replace('+','')
  dados ['cpf_rank'] = dados ['cpf_rank'].str.replace('+','')
  dados ['ifr_rank'] = dados ['ifr_rank'].str.replace('+','')
  dados ['isr_rank'] = dados ['isr_rank'].str.replace('+','')
  dados ['ger_rank'] = dados ['ger_rank'].str.replace('+','')
  dados ['irn_rank'] = dados ['irn_rank'].str.replace('+','')
  return dados


```
